```python
import requests
from prettytable import PrettyTable
import pandas as pd
import json
```

```python
obj=PrettyTable()
key='1d22c6a9-b3ac-45a3-8e47-7768db3a09ef'
```

```python
api_endpoint = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?CMC_PRO_API_KEY='
api_endpoint+=key
api_endpoint
```

```python
json_data=requests.get(api_endpoint).json()
```

```python
crypto_data=json_data['data']
```

```python
crypto_data
```

```python
for crypto in crypto_data:
    name=crypto['name']
    symbol = crypto['symbol']
    slug=crypto['slug']
    price = crypto['quote']['USD']['price']
    percent_change_1h = crypto['quote']['USD']['percent_change_1h']
    percent_change_24h = crypto['quote']['USD']['percent_change_24h']
    percent_change_7d = crypto['quote']['USD']['percent_change_7d']
    market_cap = crypto['quote']['USD']['market_cap']
    obj.add_row([name,symbol,slug,price,percent_change_1h,percent_change_24h,percent_change_7d,market_cap])
```

```python
obj.field_names=['Name','Symbol','Slug','Price','Percent_change_1h','Percent_change_24h','Percent_change_7d','market_cap']
```

```python
print(obj)
```

```python
txt=obj.get_string()
with open('output.txt','w') as file:
    file.write(txt)
```
